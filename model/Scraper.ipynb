# Pokemon Image Scraper
An image scraper implementation using BeautifulSoup. This notebook includes all the necessary code for downloading the pokemon sprites.

## Imports

```python
import requests
from bs4 import BeautifulSoup
from os import path
from urllib.parse import urljoin
```

## Constants

```python
BASE_URL = "https://pokemondb.net/sprites"
```

```python
request = requests.get(BASE_URL)
html = request.text
soup = BeautifulSoup(html, 'html.parser')
```

```python
spriteLinks = [a for a in soup.find_all("a") if a.get("class") == ["infocard"]][:890]
```

## Download

```python
for i,spriteLink in enumerate(spriteLinks):
    span = spriteLink.findChildren("span")
    img = spriteLink.findChildren("img")
    
    if span:
        imgLink = span[0].get("data-src")
    elif img:
        imgLink = img[0].get("src")

    img = requests.get(imgLink)
    filename = imgLink.split("/")[-1]
    destination = path.join("images", "swsh", filename)

    with open(destination, "wb") as f:
        f.write(img.content)
    print(f"{i+1}. Wrote {destination}")
```
